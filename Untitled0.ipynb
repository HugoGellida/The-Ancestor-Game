# The Ancestor Game

- Jeu Tour par Tour
- Objectif: capturer le plus de zones possibles
- Par Joueur:
  - Placer un drapeau sur une zone
    - La zone devient capturable
    - Une zone peut avoir 2 drapeaux
  - Ajouter une flèche
    - La flèche peut provoquer un combo
    - Une flèche est reliée entre une zone et une autre
  - Capture simple
    - Le joueur capture une zone seule
  - Capture combo
    - Le joueur choisi 2 zones
    - Toutes les zones non-capturées jusqu'à l'ancêtre commun le plus récent des 2 zones sont capturées


Une zone capturée ne peut être recapturée.

## Drapeaux

```python
class Flag:
  def __init__(self, pid):
    self.player_id = pid
```

## Node en général

```python
class Node:
  def __init__(self, id):
    self.id = id

  def set_parent(self, node):
    self.parent = node
```

## Node du jeu

```python
class GameNode(Node):
  def __init__(self, id):
    super().__init__(id)
    self.flags = []
    self.capture = False

  def add_flag(self, flag):
    if not self.capture: self.flags.append(flag)

  def get_flags(self):
    return self.flags

  def is_captured(self):
    return self.capture

  def has_flag_from_player(self, pid):
    for f in self.flags:
      if f.player_id == pid: return True
    return False
```

## Player

```python
class Player:

  def __init__(self, pid):
    self.player_id = pid
    self.zones_captured = 0;

  def can_capture(self, n1, n2):
    if (n1 == n2 and not n1.is_captured()):
      return True

    # Script des ancetres ici
    return True

  def capture(self, n1, n2):
    if (self.can_capture(n1, n2)):
      # Do it and add to zones_captured
      return True

  def place_flag(self, n):
    if (n.has_flag_from_player(self.player_id)): return
    n.add_flag(Flag(self.player_id));

  def give_parent(self, n_child, n_parent):
    n_child.set_parent(n_parent)
```

## Game

```python
import random

class Game:
  def __init__(self, p1, p2, node_amount):
    self.players = [p1, p2]
    self.nodes = []
    for i in range(0, node_amount):
      self.nodes.append(GameNode(i))

  def launch(self):
    # ===== CHOOSE RANDOM PLAYER =====
    current_player = random.choice(self.players)

    # ===== GAMEPLAY LOOP =====
    while True:  # While no players have atleast 50% zones captured

      for p in self.players:
        if p.zones_captured >= len(self.nodes) / 2:
          print("Player", p.player_id, "wins!")
          return

      # ===== PLAYER TURN =====

      n = random.choice(self.nodes)
      current_player.place_flag(n)

      if current_player.capture(n, n):
        n.setCapture(True)
        current_player.zones_captured += 1

      # ===== SWITCH PLAYER =====
      if current_player == self.players[0]:
        current_player = self.players[1]
      else:
        current_player = self.players[0]
```
